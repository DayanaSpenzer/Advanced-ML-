```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
def split_scalar(in_x,out_y):
    x_train,x_test,y_train,y_test=train_test_split(in_x,out_y,test_size=0.25,random_state=0)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def decision(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier()
    classifier.fit(x_train,y_train)
    importance=classifier.feature_importances_
    #for i, v in enumerate(importance):
      #  print(f'Feature: {i}, Score: {v:.5f}')
      #  plt.bar([x for x in range(len(importance))], importance)
      #  plt.show()
    return classifier,importance   
```

```python
def ranforest(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier()
    classifier.fit(x_train,y_train)
    importance=classifier.feature_importances_
    #for i, v in enumerate(importance):
      #  print(f'Feature: {i}, Score: {v:.5f}')
      #  plt.bar([x for x in range(len(importance))], importance)
      #  plt.show()
    return classifier,importance   
```

```python
def ranforest_permutation(x_train,y_train,x_test):
    from sklearn.inspection import permutation_importance
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.datasets import make_classification
    from sklearn.model_selection import train_test_split
    classifier = RandomForestClassifier()
    classifier.fit(x_train, y_train)

# Perform permutation importance
    results = permutation_importance(classifier, x_test, y_test, scoring='accuracy')
    importance=results.importances_mean

    return classifier, importance
```

```python
dataset1=pd.read_csv("prep.csv",index_col=None)

df2=dataset1
df2=pd.get_dummies(df2,drop_first=True)

in_x=df2.drop('classification_yes', axis=1)
out_y=df2['classification_yes']
```

```python
x_train,x_test,y_train,y_test=split_scalar(in_x,out_y)
scoredes=[]
scorerf=[]
scorerfp=[]
classifier,importance=decision(x_train,y_train,x_test)
scoredes.append(importance)
classifier,importance=ranforest(x_train,y_train,x_test)
scorerf.append(importance)
classifier,importance=ranforest(x_train,y_train,x_test)
scorerfp.append(importance)
result=Feature_Imp_Classification(classifier,importance)
result
```
